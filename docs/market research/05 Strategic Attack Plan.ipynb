## Key Business Terms in This Document

### Revenue & Funding

| Term | Meaning |
| --- | --- |
| **ARR** | Annual Recurring Revenue — predictable yearly subscription income. THE metric for SaaS companies. |
| **Pilot** | A paid trial engagement where a customer uses the product on a limited scope (e.g., one building) to evaluate before committing to a full campus. |
| **Logo** | Industry slang for "customer." "Adding logos" = signing new customers. |
| **ACV** | Annual Contract Value — what one customer pays per year. |
| **Bootstrap** | Self-funding a company from revenue (no external investors). Slower growth but you own 100%. |
| **Series A/B/C** | Funding rounds. A = first institutional raise (\~$5–20M), B = growth (\~$20–50M), C = scale (\~$50–150M). |
| **Strategic investor** | Someone who invests AND is also a customer or partner (e.g., Intel Capital investing in Buildots then also USING Buildots in their fabs). Best kind of investor for enterprise SaaS. |
| **LP** | Limited Partner — the people who put money into venture capital funds. A "strategic LP" is an industry player that can also bring deals. |

### Sales & Go-to-Market

| Term | Meaning |
| --- | --- |
| **GTM** | Go-to-Market — how you sell your product (direct sales, channel partners, marketplace, etc.) |
| **Warm intro** | An introduction to a potential buyer made by someone they already trust (vs. cold emailing a stranger). 10x more effective. |
| **Champion** | An internal advocate at the buyer's organization who pushes to purchase your software. You need a champion to close enterprise deals. |
| **Displacement** | Replacing a competitor's product at a customer. Harder than a new sale because the customer has to migrate. |
| **Channel partner** | A company that sells or recommends your product to THEIR clients. HH Angus recommending CXPro to their hospital clients = channel. |
| **VOR** | Vendor of Record (Ontario) — pre-approved supplier list. Once on it, government buyers can purchase without new procurement. |
| **Procore Marketplace** | Procore's app store where 16,000+ construction companies discover and install integrations. Being listed = distribution + legitimacy. |

### Risk & Competition

| Term | Meaning |
| --- | --- |
| **Moat** | What prevents competitors from copying you. A wider moat = more defensible business. |
| **Window** | The time period during which a market opportunity is open before competitors close it. "The window is closing" = move faster. |
| **IP** | Intellectual Property — patents, proprietary data, trade secrets that have legal protection. |
| **Provisional patent** | A preliminary patent filing (cheaper, faster) that gives you 12 months of "patent pending" status while you file the full version. |
