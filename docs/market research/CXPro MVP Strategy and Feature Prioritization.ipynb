### What is Commissioning (Cx)?

**Commissioning** is the systematic process of verifying that a new building's systems (HVAC, electrical, plumbing, fire protection, controls) actually work as designed before the owner takes possession. Think of it as QA for buildings.

When you build a $500M data center, you don't just turn the key and hope. You methodically test every piece of equipment, every system integration, and every failure scenario. That process is commissioning.

**Why it matters financially:** A hyperscale data center earns revenue the moment it's operational. Every week the commissioning process takes longer than necessary is a week of lost revenue. For a 60 MW facility, that cost-of-delay is approximately **$14.2M per month** — $3.5M per week.

### The Commissioning Levels (L0–L5) Explained Simply

Think of commissioning as a ladder — you can't skip rungs:

| Level | Plain English | Analogy |
| --- | --- | --- |
| **L0 — Design Review** | "Did we design it right on paper?" | Reviewing the blueprint before building |
| **L1 — Factory Witness** | "Did the manufacturer build the equipment correctly?" | Watching your custom furniture being built at the factory |
| **L2 — Installation Verification** | "Did the contractor install it correctly?" | Checking the furniture is assembled and placed where you specified |
| **L3 — Component Functional** | "Does each piece of equipment work on its own?" | Turning on each appliance individually to confirm it powers up |
| **L4 — System Functional** | "Do the pieces work together as a system?" | Running the whole kitchen — oven, hood vent, fire suppression — simultaneously |
| **L5 — Integrated Systems** | "Does the whole building work under stress?" | Simulating a power outage to verify backup generators, UPS, and cooling all respond correctly together |

**The gating rule:** You cannot start L3 testing until L2 is complete for that equipment. You cannot start L4 until all L3 tests in that system pass. This prevents wasted effort — testing a chiller (L3) before confirming it's wired correctly (L2) produces false failures.

### The Gap We're Filling

**The current state (how commissioning works today on a hyperscaler project):**

1. A CxA manually reads the OPR, BOD, and equipment submittals (hundreds of PDFs)
2. They manually write test procedures in Word/Excel — one per piece of equipment, 2,000–5,000 total
3. Field techs execute tests with paper/tablet checklists, log results in CxAlloy or spreadsheets
4. When tests fail, the CxA manually reads the deviation, figures out whose fault it is, and emails the responsible sub
5. At project end, the CxA manually hunts through emails, SharePoint, and Procore to assemble the turnover package

**What's broken:**

| Problem | Impact | Root Cause |
| --- | --- | --- |
| Writing 5,000 test procedures takes 6–8 weeks | Delays the entire Cx schedule | No tool generates procedures from design docs |
| 50–200 deviations/day overwhelm the CxA | Senior engineers waste 2–4 hrs/day on triage instead of solving problems | No tool classifies and routes deviations automatically |
| Incumbent software breaks at 1,000+ assets | Hyperscalers need 10,000+ assets per campus | CxAlloy/BlueRithm were built for single buildings |
| Turnover dossier takes 2–6 weeks to compile per building | Owner can't start operating (and earning revenue) | Documents scattered across 5+ systems with no structured linkage |
| No traceability from test → code citation | Compliance risk on regulated projects (pharma, hospital, federal) | Nobody has mapped test procedures to regulatory requirements |

**What CXPro does differently:**

Instead of being a better checklist app (what all incumbents are), CXPro is an **agent system** that:
- **Generates** the work (procedures from design docs) instead of just tracking it
- **Triages** failures in real-time instead of waiting for a human to read and route them
- **Compiles** the output (turnover dossier) automatically instead of manually
- **Scales** to 10,000+ assets because it was architected for that from day one
- **Links** every action to its regulatory justification (compliance copilot)

The difference in one sentence: **CxAlloy tracks what humans do. CXPro does the work and lets humans review.**

### Feature 1: Structured Commissioning Data Model (L0–L5)

### Feature 2: Test Procedure Generator Agent

**What it is:** An AI agent that ingests design documents (OPR, BOD, equipment submittals, control sequences) and generates draft Level 3/4/5 test procedures with structured steps, expected results, and acceptance criteria.

**How it works:**

```
INPUT:
- Equipment submittal PDF (e.g., Trane RTAC 500-ton chiller)
- Sequence of Operations from controls spec
- Owner's Project Requirements (performance criteria)

AGENT PROCESS:
1. Extract equipment parameters (capacity, setpoints, safeties, alarms)
2. Map to standard test categories (startup, normal operation, failover, shutdown)
3. Generate test steps with measurable acceptance criteria
4. Cross-reference against ASHRAE 202 required tests for equipment type
5. Flag gaps ("submittal mentions VFD but no VFD test generated — review")

OUTPUT:
- Structured test procedure document (not a PDF — a data object in the schema)
- Each step: action, expected result, measurement method, tolerance, pass/fail criteria
- Linked to the specific Asset and System in the data model
- Draft status — requires CxA review before execution
```

**Example output for a chiller L3 test:**

| Step | Action | Expected Result | Criteria | Method |
| --- | --- | --- | --- | --- |
| 1 | Command chiller START from BAS | Chiller starts within 30 seconds | Start confirmed via BAS point CHR-01.START_FB | BAS trend log |
| 2 | Allow chiller to reach steady state | LCHWT reaches 42°F ±2°F | Stable for 5 min | Temp sensor CHR-01.LCHWT |
| 3 | Reduce load to 25% | Chiller unloads to 25% | Capacity 23–27% | BAS point CHR-01.%RLA |
| 4 | Simulate low-flow safety | Chiller trips on low flow | Trip within 10 sec | Close isolation valve, observe |
| 5 | Reset safety, command restart | Chiller restarts normally | Same as Step 1 | BAS command |

**Why this is highest value:**
- Writing test procedures is 30–40% of a Cx agent's billable time on a hyperscaler project
- A 60 MW data center has 2,000–5,000 assets, each needing L3 at minimum — that's 2,000–5,000 procedures
- If the agent generates 80% accuracy drafts that need 20% human review, you've compressed weeks of procedure writing into days
- **The ROI pitch to a hyperscaler:** "Your Cx team spends 6 weeks writing procedures. We do it in 3 days. At $14.2M/month cost-of-delay, even 1 week saved = $3.5M in value."
- This is the demo that closes the pilot

**Build time estimate:** 3–4 weeks (ingestion pipeline + prompt engineering + structured output formatting)

### Feature 3: Deviation Triage Agent

**What it is:** When a test fails, the agent automatically classifies the deviation, assesses severity, identifies the likely responsible party (which subcontractor), and routes it for resolution.

**How it works:**

```
INPUT:
- Failed test step (from Feature 1 schema)
- Deviation description (technician's field note)
- Asset context (equipment type, system, location)
- Project context (which subs are on this system, contract scopes)

AGENT PROCESS:
1. Classify deviation type:
   - Installation deficiency (wrong, missing, damaged)
   - Controls issue (sequence error, point mapping, network)
   - Design issue (spec doesn't match OPR)
   - Equipment defect (manufacturer warranty)
   - Coordination issue (trade interference)
2. Assess severity: Critical (safety/code) / Major (function impaired) / Minor (cosmetic/documentation)
3. Identify responsible party based on contract scope matrix
4. Generate recommended resolution + re-test requirements
5. Route to responsible party with deadline based on severity

OUTPUT:
- Structured deviation record (in the schema)
- Auto-populated fields: type, severity, responsible sub, target resolution date
- Suggested resolution path
- Re-test procedure (often a subset of the original test)
```

**Why this is highest value:**
- On a hyperscaler project with 1,500+ field crew, deviations pile up at 50–200/day during peak testing
- Manual triage = a senior CxA spending 2–4 hours/day just reading and routing issues
- Misrouted deviations (sent to wrong sub) add 3–7 days of delay per issue
- **The ROI pitch:** "200 deviations/day × 15 min manual triage = 50 hours/day of senior CxA time. Our agent does it in real-time with 90% routing accuracy."

**Build time estimate:** 2–3 weeks (classification model + routing logic + notification system)
